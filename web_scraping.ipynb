```python
from bs4 import BeautifulSoup 
import requests
```

```python
url='https://en.wikipedia.org/wiki/List_of_India_Twenty20_International_cricketers'
```

```python
page=requests.get(url)
```

```python
soup=BeautifulSoup(page.text,'html')
```

```python
table=soup.find_all('table')[2]
```

```python
thead=table.find_all('th',scope='col')
thead
```

```python
headings=[head.text.strip() for head in thead]
headings
```

```python
first_col=table.find_all('th',scope='row')
first_col
```

```python
import pandas as pd
```

```python
cricketers=pd.DataFrame(columns=headings)
cricketers
```

```python
name_col=[name.text.strip() for name in first_col]
name_col
```

```python
cricketers.drop('No.',axis=1)
```

```python
cricketers.Name=name_col
temp=pd.DataFrame(columns=cricketers.columns[1:])
temp.drop('Name',axis=1)
```

```python
data=table.find_all('tr')
```

```python
for row in data[1:]:
    r=row.find_all('td')
    each_row=[i.text.strip() for i in r]
    # print(each_row)
    leng=len(temp)
    temp.loc[leng]=each_row
```

```python
temp['Name']=cricketers['Name']
```
